```python
Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.
```

```python
from flask import Flask
app = Flask(__name__)
@app.route("/")
def hello_world():
    return "Hello World!!"
if __name__ == '__main__':
    app.run(host = "0.0.0.0")
```

```python
Q3. What is App routing in Flask? Why do we use app routes?
```

```python
App routing refers to the process of binding URLs to functions within a Flask application. When a user
makes a request to a particular URL, the Flask application checks if there is a matching route defined 
for that URL, and if so, it executes the view function associated with that route.
```

```python
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
```

```python
from flask import Flask
app = Flask(__name__)
@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"
@app.route("/")
def company_details():
    return "Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999"
if __name__ == '__main__':
    app.run(host = "0.0.0.0")
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

```python
In Flask, the url_for() function is used for URL building.
It generates a URL for the specified endpoint, which is a named function that is decorated with the @app.route() decorator.
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return f'Home Page'

@app.route('/about')
def about():
    return f'About Page'

@app.route('/contact')
def contact():
    return f'Contact Page'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index', _external=True))
        print(url_for('about', _external=True))
        print(url_for('contact', _external=True))
    app.run(host = "0.0.0.0")
```
